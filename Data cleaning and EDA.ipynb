# data cleaning

```python
#fill in the missing value
age_mean = round(train_df["Age"].mean())
train_df["Age"].fillna(age_mean, inplace=True)
```

```python
train_df["Cabin"].mode()[0]
```

```python
train_df.isna().sum()
```

```python
# fill the missing values in embark with mode
Embarked_mode= train_df["Embarked"].mode()[0]
train_df["Embarked"].fillna(Embarked_mode, inplace=True)
```

```python
train_df
```

```python
extract = lambda x: x.split()[1].strip(".")#funtion to extract title from the name column
#add title to the data frame 
```

```python
train_df["Name"].apply(extract)
```

```python
#add title to the data frame
train_df["Title"]=train_df["Name"].apply(extract)
```

```python
train_df
```

```python
major_title = ["Mr", "Mrs", "Miss", "master", "Dr" "Rev"]
for i in train_df["Title"]:
    if i not in major_title:
        train_df["Title"].replace(i, "other", inplace=True)
```

```python
train_df["Title"].value_counts()
```

```python
#Data  exploration
univariance analysis on category columns
```

```python
survive_count=train_df["Survived"].value_counts()  # make a count of survive
survive_count.index = ["Dead", "Alive"]   # change the index
survive_count
```

# illustration on how to plot with pandas

# illustration on how to plot using seaborn

```python
sns.barplot(x=survive_count.index, y=survive_count.values)
```

```python
sns.set_palette("Set3")
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(x=survive_count.index, height = survive_count.values)
ax.grid(axis="y")
ax.set_facecolor("grey")
plt.title("count of survived and non survive")
plt.xticks(rotation = 30, color="green")
for idx, val in enumerate(survive_count.values):
    plt.text(x=idx, y=val-20, s=val, va="bottom", ha="center", color="purple")
plt.show()
```
